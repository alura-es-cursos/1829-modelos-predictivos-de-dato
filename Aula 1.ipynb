```python
def ip_por_pais(ip):
    try:
        return ip_df.pais[(ip_df.limite_inferior_ip < ip) & (ip_df.limite_superior_ip > ip)].iloc[0]
    except IndexError:
        return "Desconocido"
```

```python
fraude_df['ip_pais'] = fraude_df.ip.apply(ip_por_pais)
```
