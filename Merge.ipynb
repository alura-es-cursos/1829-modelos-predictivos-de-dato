```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
   'id':[1,2,3,4,5,6],
   'Nombre': ['Alex', 'Beatriz', 'David', 'Maria', 'Paula', 'Rodrigo'],
   'Matematicas':[4.5, 5.0, 3.2, 2.8, 4.1, 3.9]})

df2 = pd.DataFrame(
   {'id':[1,2,4,6,7,8,9],
   'Nombre': ['Alex', 'Beatriz', 'Maria', 'Rodrigo', 'Fernando', 'Natalia', 'Maria'],
   'Literatura':[4.8, 4.7, 3.0, 3.8, 4.0, 4.3, 5.0]})
```

```python
df1
```

```python
df2
```

## Funci√≥n Merge con base en llaves

#### Merge usando una llave: Id

```python
pd.merge(df1, df2, on='id')
```

#### Merge usando una llave: Nombre

```python
pd.merge(df1, df2, on='Nombre')
```

#### Merge usando dos llaves

```python
pd.merge(df1, df2, on=['id','Nombre'])
```

## Merge usando el argumento 'how'

#### Left Join 

```python
pd.merge(df1, df2, on='id', how='left')
```

#### Right Join 

```python
pd.merge(df1, df2, on='id', how='right')
```

#### Outer Join

```python
pd.merge(df1, df2, on='id', how='outer')
```

#### Inner Join 

```python
pd.merge(df1, df2, on='id', how='inner')
```
